**Q.1.** Given a non-negative integer x, return the square root of x rounded down to the nearest integer. The returned integer should be non-negative as well. You must not use any built-in exponent function or operator. 

**Example 1:**

```
Input: x = 4
Output: 2
Explanation: The square root of 4 is 2, so we return 2.
```

**Example 2:**

```
Input: x = 8
Output: 2
Explanation: The square root of 8 is 2.82842..., and since we round it down to the nearest integer, 2 is returned.
Constraints:

0 <= x <= 2^31 - 1
```

**Note:** Create a GitHub file for the solution and add the file link the the answer section below.

````python
# Define a function to return the square root of x rounded down to the nearest integer
def sqrt(x):
    # Check if x is non-negative
    if x < 0:
        return "Invalid input"
    # Check if x is zero or one
    if x == 0 or x == 1:
        return x
    # Initialize low and high variables for binary search
    low = 0
    high = x
    # Loop until low is less than or equal to high
    while low <= high:
        # Find the mid point
        mid = (low + high) // 2
        # Check if mid is the square root of x
        if mid * mid == x:
            return mid
        # If mid is too small, move low to the right of mid
        elif mid * mid < x:
            low = mid + 1
            # Update the answer to the closest integer below the square root
            ans = mid
        # If mid is too large, move high to the left of mid
        else:
            high = mid - 1
    # Return the answer
    return ans

# Test the function with some examples
print(sqrt(4)) # Output: 2
print(sqrt(8)) # Output: 2
````

**Q.2.** You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.


**Example 1:**

```
Input: l1 = [2,4,3], l2 = [5,6,4]
Output: [7,0,8]
Explanation: 342 + 465 = 807.
```

**Example 2:**

```
Input: l1 = [0], l2 = [0]
Output: [0]
```

**Example 3:**

```
Input: l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]
Output: [8,9,9,9,0,0,0,1]
```

**Constraints:**

The number of nodes in each linked list is in the range [1, 100].

0 <= Node.val <= 9 It is guaranteed that the list represents a number that does not have leading zeros.

**Note:** Create a GitHub file for the solution and add the file link the the answer section below.

````python
# Define a class for the linked list nodes
class ListNode:
  # Initialize the node with a value and a next pointer
  def __init__(self, val=0, next=None):
    self.val = val
    self.next = next

# Define a function to add two numbers represented by linked lists
def addTwoNumbers(l1, l2):
  # Initialize a dummy head for the result list
  dummy = ListNode()
  # Initialize a current pointer to the dummy head
  curr = dummy
  # Initialize a carry variable to store the carry over from each addition
  carry = 0
  # Loop until both lists are exhausted or carry is non-zero
  while l1 or l2 or carry:
    # Get the values of the current nodes of l1 and l2, or 0 if they are None
    x = l1.val if l1 else 0
    y = l2.val if l2 else 0
    # Add the values and the carry and get the sum and the new carry
    sum = x + y + carry
    carry = sum // 10
    # Create a new node with the value of sum modulo 10 and append it to the result list
    curr.next = ListNode(sum % 10)
    curr = curr.next
    # Move l1 and l2 to the next nodes if they exist
    if l1:
        l1 = l1.next
    if l2:
        l2 = l2.next
  # Return the result list without the dummy head
  return dummy.next

# Define a helper function to convert a list to a linked list
def listToLinkedList(lst):
  # Initialize a dummy head for the linked list
  dummy = ListNode()
  # Initialize a current pointer to the dummy head
  curr = dummy
  # Loop through the list and create nodes with the values and append them to the linked list
  for val in lst:
    curr.next = ListNode(val)
    curr = curr.next
  # Return the linked list without the dummy head
  return dummy.next

# Define a helper function to convert a linked list to a list
def linkedListToList(head):
  # Initialize an empty list to store the values of the linked list nodes
  lst = []
  # Initialize a current pointer to the head of the linked list
  curr = head
  # Loop until the current pointer is None and append the values to the list
  while curr:
    lst.append(curr.val)
    curr = curr.next
  # Return the list
  return lst

# Test the function with some examples
l1 = [2,4,3]
l2 = [5,6,4]
print(linkedListToList(addTwoNumbers(listToLinkedList(l1), listToLinkedList(l2)))) # Output: [7,0,8]

l1 = [0]
l2 = [0]
print(linkedListToList(addTwoNumbers(listToLinkedList(l1), listToLinkedList(l2)))) # Output: [0]

l1 = [9,9,9,9,9,9,9]
l2 = [9,9,9,9]
print(linkedListToList(addTwoNumbers(listToLinkedList(l1), listToLinkedList(l2)))) # Output: [8,9,9,9,0,0,0,1]
````
